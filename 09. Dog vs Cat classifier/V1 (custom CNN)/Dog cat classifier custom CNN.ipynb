## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, BatchNormalization, Flatten, Dropout
```

## Importing dataset

### Training set

```python
img_gen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)

train_set_generator = img_gen.flow_from_directory(
    '/content/drive/MyDrive/Self/22. Deep Learning/02. CNN/Deep Learning AZ/Convolutional_Neural_Networks 3/dataset/training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary',
)
```

## Test set

```python
img_test_gen = ImageDataGenerator(
    rescale=1./255
)

test_set_generator = img_test_gen.flow_from_directory(
    '/content/drive/MyDrive/Self/22. Deep Learning/02. CNN/Deep Learning AZ/Convolutional_Neural_Networks 3/dataset/test_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary',
)
```

## Design model

```python
model1 = Sequential([
      Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu', input_shape=[64, 64, 3]),
      MaxPool2D(pool_size=2, padding='valid'),
      Conv2D(filters=128, kernel_size=(3, 3), padding='same', activation='relu'),
      MaxPool2D(pool_size=2, padding='valid'),

      Conv2D(filters=256, kernel_size=(3, 3), padding='same', activation='relu'),
      MaxPool2D(pool_size=2, padding='valid'),

      Flatten(),
      Dense(256, activation='relu'),
      Dense(1, activation='sigmoid')
])
```

### Compilling model

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Training model 1

```python
history = model1.fit(train_set_generator, validation_data=test_set_generator, batch_size=64,epochs=50, verbose=1)
```

```python
model1.evaluate(test_set_generator)
```
